## Usandod VGG16 para Classificar Porcos

```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
```

```python
modelo = VGG16(weights='imagenet')
```

```python
imagem   = "0,,43969382,00.jpg"
target   =  image.load_img(imagem, target_size=(224, 224))
img_data = image.img_to_array(target)
img_data = np.expand_dims(img_data, axis=0)
img_data = preprocess_input(img_data)
```

```python
preds = modelo.predict(img_data)
for p in decode_predictions(preds, top=5)[0]:
  print(p[1])
  
```
